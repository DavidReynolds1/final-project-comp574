# Project Milestone 1 - Getting Data

```python
# As of right now these are the only two locations that I know will work and are populated fully
# All location must be within the United States - I convert US zipcodes later on
# 03301 = Zip code for Concord, NH
# 03060 = Zip Code for Nashua, NH
import requests, json, pandas as pd

api_key = "8390c20af9c3e9f5dcf4f89d947eba14"
base_url = "http://api.openweathermap.org/data/2.5/forecast?"
zips=["03301", "03060"]

# For loop to split up different zips as I want two separate dataframes. Probably not the best way, but functional.
for zip in zips:
    if zip == "03301":
        complete_url = base_url + "zip=" + zip + "&appid=" + api_key
    else:
        complete_url1 = base_url + "zip=" + zip + "&appid=" + api_key
```

```python
# creates json payload for the urls made above
payload = requests.get(complete_url).json()
payload1 = requests.get(complete_url1).json()
```

```python
# Creates dataframes from the payload --- normalize = json --> dataframe
df = pd.json_normalize(payload["list"]).drop("main.temp_kf", axis= 1)
df1 = pd.json_normalize(payload1["list"]).drop("main.temp_kf", axis= 1)
```
